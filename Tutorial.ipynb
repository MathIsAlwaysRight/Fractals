<h1>2 Coloring and resizing</h1>
<h3>2.1 Zooming</h3>
<p>Let's get back to the mandelbrot set, and let's try to zoom in. For zooming, we will use xRange, yRange and axes parameter.</p>

```python
mandelbrot=Fractal(lambda z,c: z**2+c)
mandelbrot.drawImage(xRange=[-1.5,0.5], yRange=[-1,1], axes=True) #Now we can see what the actual complex numbers are (Thanks to axes=True)
```

```python
mandelbrot=Fractal(lambda z,c: z**2+c)
mandelbrot.drawImage(xRange=[-0.75,-0.5], yRange=[-0.75,-0.5], axes=True) #Thanks to axes we could zooom on the exact location we wanted
```

```python
mandelbrot=Fractal(lambda z,c: z**2+c, maxIter=300) #For further zooms, increase the maxIter parameter
mandelbrot.drawImage(xRange=[-0.6,-0.59], yRange=[-0.625,-0.615], axes=True) #We found little deformed copy of original mandelbrot :)
```

<h3>2.2 Coloring</h3>
<p>Let's move to coloring now</p>

```python
mandelbrot=Fractal(lambda z,c: z**2+c)
mandelbrot.drawImage(xRange=[-1.5,0.5], yRange=[-1,1], hsvMap=[lambda iter: iter, 1, 1]) #Into hsvMap, put the [hue, saturation, brightness]
#So what this [lambda iter: iter, 1, 1] thing is saying, is that color HUE will be dependent on iterations, saturation will be 1 (full) and brightness 1 (Brightest)
```

```python
mandelbrot=Fractal(lambda z,c: z**2+c)
mandelbrot.drawImage(xRange=[-1.5,0.5], yRange=[-1,1], hsvMap=[0.6, lambda iter: iter*2, 1]) #Here is image dependent on saturation
```

<h3>2.3 Coloring tricks</h3>
<p>We will just briefly touch all the posibilities you can do with coloring, there are lots more of course</p>

```python
mandelbrot=Fractal(lambda z,c: z**2+c)
mandelbrot.drawImage(xRange=[-1.5,0.5], yRange=[-1,1], hsvMap=[0.6, lambda iter: (iter*100)%2, 1]) #Let's try something really weird. 
#Modulo 2 (%2) returns the remainder after division by 2. So 1%2=1, 4%2=0, 7%2=1 etc... We gotta multiply iter by 100, because iter is a float between 0 and 1, and 100 is our maxIter
```

```python
mandelbrot=Fractal(lambda z,c: z**2+c)
mandelbrot.drawImage(xRange=[-1.5,0.5], yRange=[-1,1], hsvMap=[lambda iter: ((iter*100)%5)/5, 1, 1], hsvCvg=[0,0,1]) #And rainbow-brot, we also changed hsvCvg to 0, 0, 1
#So the point that actually belong into the mandelbrot set, will be hue 0 saturation 0 and brightness 1. Which is the white color
```

<h3>3.4 saveGif</h3>
<p>You can save gifs using this function. Unlike other functions it's bit complicated.</p>

```python
Fractal.saveGif(lambda z,c,i: (3**(-i-0.2)-2.5)**z, fBounds=[-1,2], frames=40, FPS=40, maxIter=75, borderRad=50, size=[550,550], hsvMap=[lambda iter: iter*0.75+0.5, 1, lambda iter: 0.7-iter*1.5])
#The first argument is a lambda function, you already know z and c, and i is the iterating parameter, it's what's changing from image to image
#Fbounds are bounds of the parameter i. In this case, i will go from -1 to 2. 
#frames is number of frames, more frames take more time but it's more smooth
#FPS is frames per seconds
```

<h3>3.5 Convergence Check and pic</h3>
<p>Using this, you can focus on a single point and check if, and how it converges.</p>

```python
mandelbrot=Fractal(lambda z,c: z**2+c) #This is the basic mandelbrot
print(mandelbrot.ConvergenceCheck(0+1j)) #You can see that the imaginary unit i converges, so it belongs in the mandelbrot set
print(mandelbrot.ConvergenceCheck(1)) #While point 1 diverges actually pretty quickly. (Closer to 0 --> Diverges quickly)
```

```python
mandelbrot.drawImage(xRange=[-1.5,0.5], yRange=[-1,1], axes=True)
```

```python
mandelbrot.ConvergencePic(-0.4+0.4j, steps=50) #Here you can see point -0.4+0.4i converging into value around -0.35+0.22j
#Every line symbolizes 1 step. So it went from initial position -0.4+0.4i to -0.4+0.5i all the way to around -0.35+0.22j (X axis is real and Y axis is imaginary)
```
